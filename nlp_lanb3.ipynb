```python
class trie_suff:
    def __init__(self) -> None:
        self.links = {}
        self.freq = 0

    def insert(self, x: str):
        p = self
        for i in reversed(x):
            if i not in p.links:
                p.links[i] = trie_suff()
            p.freq += 1
            p = p.links[i]

    def stem(self, x: str) -> str:
        p = self
        ans, ans2 = '', ''
        block_size = 5
        for i in reversed(x):
            ans2 = i + ans2
            if i not in p.links:
                break
            if p.links[i].freq > block_size:
                p = p.links[i]
                ans = ans2
            else:
                break
        return ans

```
