```python
# Take the first 10000 lines from the dataset
sample_data = list(dt.take(1000))
```

```python
import re
#. ? !
pattern = r'^.*[.?!]$'
#now i have the pattern for it
# arr = re.findall(pattern,sample_data)
arr = []
for i in sample_data:
  arr.append(re.findall(pattern,i['text']));
print(len(arr))
# print(arr)
arr2 = []
for i in arr:
  if(len(i) != 0):
    arr2.append(i)
# print(len(arr2))
print(arr2)
```

```python
#handling the indian format data
#now we should add the custom logic for the date_reg
#writing simple reg for indian assuming we dont have any noise
date_reg = r'(\d{2}/\d{2}/\d{4})|(\d{2}-(?:\w+)-\d{4})'
#simepl date regex
date_list = []
for i in arr2:
  date_list.append(re.findall(date_reg,i[0]))
print(date_list)
# print(type(arr2[0]))
```

```python
import csv

# Flatten the list of lists `arr2` into a single list of strings
flattened_sentences = [sentence[0] for sentence in arr2 if sentence]

# Specify the output CSV file path
csv_file_path = 'sentences.csv'

# Write the sentences to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    for sentence in flattened_sentences:
        csv_writer.writerow([sentence])

print(f"Sentences saved to {csv_file_path}")
```
